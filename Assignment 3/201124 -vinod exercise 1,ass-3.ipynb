```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
customers = pd.read_csv('Ecommerce Customers')
```

```python
customers.head()
```

```python
customers.describe()
```

```python
customers.info()
```

```python
sns.set_style('whitegrid')
sns.jointplot(x="Time on Website" ,y="Yearly Amount Spent" ,data = customers , color= 'Gray')
```

```python
sns.set_style('whitegrid')
sns.jointplot(x="Time on App" ,y="Yearly Amount Spent" ,data = customers, color='black')
```

```python
sns.set_style('whitegrid')
sns.jointplot(x = 'Time on App' ,y = 'Length of Membership' , data = customers,kind = "hex",  color= "darkblue")
```

```python
sns.pairplot(customers)
```
